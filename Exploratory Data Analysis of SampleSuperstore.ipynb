# Statewise analysis¶

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State', data=dataset, palette='rocket_r',order=dataset['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

# So, here we conclude that california has highest number of order recieved and it is most sctive state for the selling purpose

# Profit and discount relation

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=dataset)
plt.legend()
plt.show()
```

# So, we can conclude that at point 5 discount profit is minimum

# Profit/loss and sales by states

```python
ds=dataset.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
ds[:].plot.bar(color=['Blue','Red'],figsize=(15,12))
plt.title("Profit/loss and sales across the state")
plt.xlabel('States')
plt.ylabel('Profit/loss and sales')
plt.show()
```

# So, if quantity is high then profit will also high

# Profit/loss and sales by Sub-category¶

```python
ds=dataset.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sub-Category', ascending=False)
ds[:].plot.bar(color=['Blue','Red'],figsize=(15,12))
plt.title("Profit/loss and sales across the Sub-category")
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss and sales')
plt.show()
```
